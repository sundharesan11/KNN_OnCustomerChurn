# K Nearest Neighbours

The k-nearest neighbors algorithm, also known as *KNN* or *k-NN*, is a non-parametric, supervised learning classifier, which uses **proximity** to make classifications or predictions about the grouping of an individual data point.

In this illustration I have explored the classification using k-NN algorithm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer, LabelEncoder, OrdinalEncoder
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix, precision_recall_fscore_support, precision_score, recall_score
```

```python
data = pd.read_csv("churndata_processed.csv")
```

```python
data.head(10)
```

```python
data.dtypes.value_counts()
```

```python
a=data.dtypes==float
a
```

```python
 
df_uniques = pd.DataFrame([[i, len(data[i].unique())] for i in data.columns], columns=['Variables', 'Unique_values']).set_index('Variables')
```

```python
df_uniques
```

```python
binary_var = list(df_uniques[df_uniques.Unique_values==2].index)
```

```python
binary_var
```

```python
cat_var = list(df_uniques[(df_uniques.Unique_values>2) & (df_uniques.Unique_values<=5)].index)
```

```python
cat_var
```

```python
[[i, list(data[i].unique())] for i in cat_var]
```

```python
ordinal_var = ['contract', 'satisfaction', 'months']
```

```python
numeric_var = list(set(data.columns)-set(ordinal_var)-set(binary_var))
```

```python
data[numeric_var]
```

```python
lb = LabelBinarizer()
le=LabelEncoder()
oe = OrdinalEncoder()
```

```python
data[ordinal_var].describe()
```

```python
for column in ordinal_var:
    data[column] = le.fit_transform(data[column])
```

```python
data[ordinal_var].astype('category').describe()
```

```python
for column in binary_var:
    data[column] = lb.fit_transform(data[column])
```

```python
cat_var = list(set(cat_var) - set(ordinal_var))
```

```python
cat_var
```

```python
data = pd.get_dummies(data, columns=cat_var, drop_first=True)
```

```python
data.describe().T
```

```python
savefile = 'churning_processed.csv'
data.to_csv(savefile, index=False)
```

```python
y, X = data['churn_value'], data.drop(columns='churn_value')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))
print('Accuracy score: ', round(accuracy_score(y_test, y_pred), 2))
print('F1 Score: ', round(f1_score(y_test, y_pred), 2))
```

```python
ax = sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')  
labels = ['False', 'True']
ax.set_xticklabels(labels, fontsize=25);
ax.set_yticklabels(labels[::-1], fontsize=25);
ax.set_ylabel('Prediction', fontsize=30);
ax.set_xlabel('Ground Truth', fontsize=30)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn = knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))
print('Accuracy score: ', round(accuracy_score(y_test, y_pred), 2))
print('F1 Score: ', round(f1_score(y_test, y_pred), 2))
```

```python
ax = sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d')  
labels = ['False', 'True']
ax.set_xticklabels(labels, fontsize=25);
ax.set_yticklabels(labels[::-1], fontsize=25);
ax.set_ylabel('Prediction', fontsize=30);
ax.set_xlabel('Ground Truth', fontsize=30)
```

```python
max_k = 40
f1_scores = list()
error_rates = list() 

for k in range(1, max_k):
    
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance')
    knn = knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    f1 = f1_score(y_pred, y_test)
    f1_scores.append((k, round(f1_score(y_test, y_pred), 4)))
    error = 1-round(accuracy_score(y_test, y_pred), 4)
    error_rates.append((k, error))
    
f1_results = pd.DataFrame(f1_scores, columns=['K', 'F1 Score'])
error_results = pd.DataFrame(error_rates, columns=['K', 'Error Rate'])
```

```python
f1_results
```

```python
error_results
```

```python
ax = f1_results.set_index('K').plot(figsize=(12,12))
ax.set(xlabel='K', ylabel='F1 Score')
ax.set_xticks(range(1, max_k));
plt.title('KNN F1 Score')
```

```python
ax = error_results.set_index('K').plot(figsize=(12,12))
ax.set(xlabel='K', ylabel='Error rate')
ax.set_xticks(range(1, max_k));
plt.title('KNN  Elbow curve')
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn = knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(classification_report(y_test, y_pred))
print('Accuracy score: ', round(accuracy_score(y_test, y_pred), 2))
print('F1 Score: ', round(f1_score(y_test, y_pred), 2))
```
